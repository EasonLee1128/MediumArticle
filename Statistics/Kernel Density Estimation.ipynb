```python
import math
import matplotlib.pyplot as plt

# 標準高斯函數
def gaussian_kernel(x):
    return (1/math.sqrt((2*math.pi))*math.exp(-0.5*x**2))

# 身高樣本
xs = [158,160,160,163,168,165,170,165,170,175,173,172,162]
```

## 計算機率密度 f(x=160) :

```python
x = 160
pdf = 0
for xi in xs:
    pdf += gaussian_kernel(x-xi)

pdf = pdf/len(xs)
print(f"x={x}, pdf={pdf}")
```

## 繪製機率密度函數折線圖

```python
total_xs = [i for i in range(min(xs),max(xs)+1,1)]
total_pdf = []

for x in total_xs:
  
    for xi in xs:
        pdf += gaussian_kernel(x-xi)
        
    pdf = pdf/len(xs)
    total_pdf.append(pdf)
    
plt.plot(total_xs,total_pdf)
plt.ylim(0)
```

## 繪製不同帶寬h的密度函數折線圖

```python
fig,ax =  plt.subplots(2,2,figsize=(10, 10))
pos_list = [(0,0),(0,1),(1,0),(1,1)]
for h in (1,2,3,4):
  
    total_xs = [i for i in range(min(xs),max(xs)+1,1)]
    total_pdf = []
    
    for x in total_xs:
      
        for xi in xs:
            pdf += gaussian_kernel((x-xi)/h)
            
        pdf = pdf/len(xs)
        total_pdf.append(pdf)
    
    r,c = pos_list[h-1]       
    ax[r][c].plot(total_xs,total_pdf) 
    ax[r][c].set_title(f"Density Function h={h}")
```
