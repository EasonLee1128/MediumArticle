# 匯入套件與基本變數定義

```python
import math
import matplotlib.pyplot as plt

# Samples & Estimated Variable
xs = [158,160,160,163,168,165,170,165,170,175,173,172,162]
total_xs = [i for i in range(min(xs),max(xs)+1,1)]
```

# 簡單估計 - 變數與樣本的平均距離

```python
# Simple estimation
total_pdf = []

for x in total_xs:
    pdf = 0
    for xi in xs:
        pdf += abs(x-xi)
        
    pdf = pdf/len(xs)
    total_pdf.append(pdf)
 
plt.title("Average Distance")
plt.plot(total_xs,total_pdf,linewidth=3)
plt.xticks(fontsize = 12)
plt.ylim(0)
```

# 核密度估計

```python
# Kernel extimation
##　Kernel function
def gaussian_kernel(x):
    return (1/math.sqrt((2*math.pi))*math.exp(-0.5*x**2))

total_pdf = []
for x in total_xs:
    pdf = 0
    for xi in xs:
        pdf += gaussian_kernel(x-xi)
        
    pdf = pdf/len(xs)
    total_pdf.append(pdf)
 
plt.title("Kernel Density Estimation")
plt.plot(total_xs,total_pdf,linewidth=3)
plt.xticks(fontsize = 12)
plt.ylim(0)
```

## 繪製不同帶寬h的密度函數折線圖

```python

fig,ax =  plt.subplots(2,2,figsize=(10, 10))
pos_list = [(0,0),(0,1),(1,0),(1,1)]
for h in (1,2,3,4):
    total_pdf = []
    for x in total_xs:
        pdf =0
        for xi in xs:
            pdf += gaussian_kernel((x-xi)/h)
            
        pdf = pdf/len(xs)
        total_pdf.append(pdf)
    
    r,c = pos_list[h-1]       
    ax[r][c].plot(total_xs,total_pdf,linewidth=3) 
    ax[r][c].set_title(f"Kernel Density Estimation h={h}")
    ax[r][c].set_title(f"Density Function h={h}")
```
